```
# Plot dos dados de entrada
def plotDados(dados):
  for i in range(0, len(dados)):
    if(saida[i] == 0):
      plt.scatter(dados[i][0], dados[i][1], marker = "o", color = "green")
    else:
      plt.scatter(dados[i][0], dados[i][1], marker = "x", color = "blue")

  plt.xlim(-0.25,1.25)
```

```
# Definir parametros iniciais do Perceptron

# Pesos das entradas 
w1 =  -0.2
w2 =  0.4

# Coeficiente Linear (bias)
b = 0

# Taxa de Aprendizagem
taxaDeAprendizagem = 0.2

# Critperio de parada
epocas = 5

fig = plt.figure()
camera = Camera(fig)

plotDados(dados)
x1 = -0.25
y1 = -1*(x1*w1+b)/w2

x2 = 1.25
y2 = -1*(x2*w1+b)/w2
plt.plot([x1, x2], [y1, y2])
camera.snap()


for geracao in range(0,epocas):
  print()
  print("Época: ",  geracao + 1)
  [w1, w2, b] = Epoca(w1,w2,b)
  plotDados(dados)
  x1 = -0.25
  y1 = -1*(x1*w1+b)/w2
  x2 = 1.25
  y2 = -1*(x2*w1+b)/w2
  plt.plot([x1, x2], [y1, y2])
  camera.snap()


animation = camera.animate()

HTML(animation.to_html5_video())
```

```
def Epoca(w1, w2, b):

  for i in range(0,len(dados)):
    # Dados de Treinamento
    print("-----Entradas-----")
    print("x1: ", dados[i][0])
    print("x2: ", dados[i][1])
    print("x1 V x2: ", saida[i])

    # Combinação Linear das entradas com os pesos
    y = (dados[i][0] * w1) + (dados[i][1] * w2)
    print("y: ", y)

    # Função Degrau
    if (y > -b):
      saidaPerceptron = 1
    else:
      saidaPerceptron = 0

    print("Degrau(Y):", saidaPerceptron)

    # Calcular Erro
    erro = saida[i] - saidaPerceptron
    print("Erro: ", erro)

    # Comparar e atualizar os pesos
    if erro != 0:
      print("-----Atualizando os Pesos-----")
      w1 = w1 + (taxaDeAprendizagem * dados[i][0] * erro)
      w2 = w2 + (taxaDeAprendizagem * dados[i][1] * erro)
      b = b + (taxaDeAprendizagem * erro)

    print("w2: ", w2)
    print("w1: ", w1)
    print("b: ", b)

  return w1, w2, b
```
